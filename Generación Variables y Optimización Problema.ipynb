```python
def data(x, y, function, clusters=8):
    np.random.seed(1)
    MC_max=2500000
    MC_min=500000
    PC_max=1000000
    #PC_min=400000
    PC_min=70000
    G_max=1.2
    G_min=0.15
    mu_o=2733480/(len(x)*len(y))
    sigma_o=30
    mu_w=8386560/(len(x)*len(y))
    sigma_w = 30
    #sigma_w=3000000
    mu_g=0.51
    sigma_g=0.3
    
    rows = []
    i, j, a, b =0, 0, 1, 1
    for n in x:
        for m in y:
            row = {
                'x':n,
                'y':m,
                'z':f(function=function, x=n, y=m, a=3),
                'o':st.norm(mu_o,sigma_o).rvs(),
                'w':st.norm(mu_w,sigma_w).rvs(),
                'g':g_value(n,m)
            }
            rows.append(row)
    df_aux=pd.DataFrame(rows)
    values = MinMaxScaler().fit_transform(df_aux)
    columns = df_aux.columns
    df = pd.DataFrame(values, columns=columns)
    
    X=df.iloc[:,[0,1,2,5]].values
    x=df.iloc[:,[0,1,2,5]].values
    
    kmeans=KMeans(n_clusters=clusters, init='k-means++', max_iter=300, n_init=10, random_state=0)
    y_kmeans=kmeans.fit_predict(x)
    
    dff=pd.concat([df_aux,pd.DataFrame(y_kmeans,columns=['Clusters'])],axis=1)
    c=list(dff.Clusters.unique())
    c.sort()
    o_sum={}
    w_sum={}
    g_mean={}

    for i in c:
        o_sum[i] = dff[dff.Clusters==i].o.sum()
        w_sum[i] = dff[dff.Clusters==i].w.sum()
        g_mean[i] = dff[dff.Clusters==i].g.mean()
        
    bloques=list(i for i in range (len(rows)))
    periodos=list(t for t in range(12))
    IT=[(i,t) for i in bloques for t in periodos]
    o={i:float(dff[dff.index==i].o) for i in range(len(rows))}
    w={i:float(dff[dff.index==i].w) for i in range(len(rows))}
    g={i:float(dff[dff.index==i].g) for i in range(len(rows))}
    r=0.12
    v={(i,t):500*o[i]*g[i]/((1+r)**t) for i in bloques for t in periodos}
    
    bloques=list(i for i in range (len(rows)))
    periodos=list(t for t in range(12))
    IT=[(i,t) for i in bloques for t in periodos]
    o={i:float(dff[dff.index==i].o) for i in range(len(rows))}
    w={i:float(dff[dff.index==i].w) for i in range(len(rows))}
    g={i:float(dff[dff.index==i].g) for i in range(len(rows))}
    r=0.12
    v={(i,t):500*o[i]*g[i]/((1+r)**t) for i in bloques for t in periodos}
    
    return bloques, periodos, IT, o, w, g, r, v, o_sum, w_sum, g_mean, c
```

```python
def optimizacion_base(bloques, periodos, IT, o, w, g, r, v):
    modelo=Model('Optimizacion base')
    modelo.setParam('LogToConsole', 0)
    x=modelo.addVars(IT,vtype=GRB.CONTINUOUS, name='x',lb=0, ub=1)
    modelo.setObjective(expr=quicksum(v[i,t]*x[i,t] for i,t in IT),sense=GRB.MAXIMIZE)
    for t in periodos: 
        modelo.addConstr(quicksum(o[i]*x[i,t] for i in bloques)>=PC_min, name=f'(1)-{t}')

    for t in periodos:
        modelo.addConstr(quicksum(o[i]*x[i,t] for i in bloques)<=PC_max, name=f'(2)-{t}')

    for t in periodos:
        modelo.addConstr(quicksum((o[i]+w[i])*x[i,t] for i in bloques)>=MC_min, name=f'(3)-{t}')

    for t in periodos:
        modelo.addConstr(quicksum((o[i]+w[i])*x[i,t] for i in bloques)<=MC_max, name=f'(4)-{t}')

    for t in periodos:
        modelo.addConstr(quicksum((o[i]*g[i])*x[i,t] for i in bloques)-quicksum(o[i]*x[i,t] for i in bloques)*G_min>=0, name=f'(5)-{t}')

    for t in periodos:
        modelo.addConstr(quicksum((o[i]*g[i])*x[i,t] for i in bloques)-quicksum(o[i]*x[i,t] for i in bloques)*G_max<=0, name=f'(6)-{t}')

    for i in bloques:
        modelo.addConstr(quicksum(x[i,t] for t in periodos)==1, name=f'(7)-{i}')
    modelo.update()
    
    modelo.optimize()
    
    if modelo.status == 3:
        print('Infactible')
        modelo.computeIIS()
        modelo.write('model.ilp')
        
    else:
        objective = modelo.objVal
        runtime = modelo.Runtime
    return objective, runtime
```

## Optimización modelo agregado

```python
def optimizacion_cluster(bloques, periodos, IT, o, w, g, r, v):
    modelo=Model('Optimizacion cluster')
    modelo.setParam('LogToConsole', 0)
    x=modelo.addVars(IT,vtype=GRB.CONTINUOUS, name='x',lb=0, ub=1)
    modelo.setObjective(expr=quicksum(v[i,t]*x[i,t] for i,t in IT),sense=GRB.MAXIMIZE)
    
    for t in periodos: 
        modelo.addConstr(quicksum(o[i]*x[i,t] for i in bloques)>=PC_min, name=f'(1)-{t}')

    for t in periodos:
        modelo.addConstr(quicksum(o[i]*x[i,t] for i in bloques)<=PC_max, name=f'(2)-{t}')

    for t in periodos:
        modelo.addConstr(quicksum((o[i]+w[i])*x[i,t] for i in bloques)>=MC_min, name=f'(3)-{t}')

    for t in periodos:
        modelo.addConstr(quicksum((o[i]+w[i])*x[i,t] for i in bloques)<=MC_max, name=f'(4)-{t}')

    for t in periodos:
        modelo.addConstr(quicksum((o[i]*g[i])*x[i,t] for i in bloques)-quicksum(o[i]*x[i,t] for i in bloques)*G_min>=0, name=f'(5)-{t}')

    for t in periodos:
        modelo.addConstr(quicksum((o[i]*g[i])*x[i,t] for i in bloques)-quicksum(o[i]*x[i,t] for i in bloques)*G_max<=0, name=f'(6)-{t}')

    for i in bloques:
        modelo.addConstr(quicksum(x[i,t] for t in periodos)==1, name=f'(7)-{i}')
    modelo.update()
    
    modelo.optimize()
    
    if modelo.status == 3:
        print('Infactible')
        modelo.computeIIS()
        modelo.write('model.ilp')
    else:
        objective = modelo.objVal
        runtime = modelo.Runtime
    
    return objective, runtime
```

```python
clusters = [4,8,10]
sizes = [10,30,80,100]
functions = ['f1', 'f2', 'f3', 'f4']
optimality_gaps = []
rows = []
for cluster in clusters:
    for size in sizes:
        for function in functions:
            bloques, periodos, IT, o, w, g, r, v, o_sum, w_sum, g_mean, c = data(x=list(np.linspace(-15,15,size)), 
                                                        y=list(np.linspace(-15,15,size)), function=function, 
                                                        clusters=cluster)
            objective_base, runtime_base = optimizacion_base(bloques=bloques, periodos=periodos, IT=IT, o=o, 
                                                             w=w, g=g, r=r, v=v)
            bloques=c.copy()

            IT=[(i,t) for i in bloques for t in periodos]
            o=o_sum.copy()
            w=w_sum.copy()
            g=g_mean.copy()
            v={(c,t):500*o[c]*g[c]/((1+r)**t) for c in bloques for t in periodos}
            #print(v)
            objective_cluster, runtime_cluster = optimizacion_cluster(bloques=bloques, periodos=periodos, 
                                                                      IT=IT, o=o, w=w, g=g, r=r, v=v)
            optimality_gap = round(abs(objective_base-objective_cluster)/objective_base*100, 2)
            optimality_gaps.append(optimality_gap)
            row = {
                'cluster':cluster,
                'size':size,
                'function':function,
                'gap':optimality_gap,
                'objective_base':objective_base,
                'objective_cluster':objective_cluster,
                'runtime_base':runtime_base,
                'runtime_cluster':runtime_cluster,
            }
            rows.append(row)
            print('\n----------------------------------------')
            print('clusters', cluster, 'size', size, 'function', function)
            print(str(optimality_gap)+'%', runtime_base, runtime_cluster)
DOE = pd.DataFrame(rows)
```
